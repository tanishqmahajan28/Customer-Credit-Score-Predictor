```python
# Checking outliers and treating them by imputting median
import matplotlib.pyplot as plt
def outlier_detection_treatment(col):
    q1= df[col].quantile(0.25)
    q3= df[col].quantile(0.75)
    iqr= q3-q1
    lf= q1- 1.5*iqr
    uf= q3+ 1.5*iqr
    outliers= df[(df[col]<lf) | (df[col]>uf)]
    print("Percentage of outliers in", col ,outliers.shape[0]*100/df.shape[0])
    fig, ax= plt.subplots(1,2, figsize=(6,4))
    ax[0].boxplot(df[col])
    ax[1].hist(df[col])
    plt.suptitle(col)
    plt.show()
    df.loc[(df[col]<lf) | (df[col]>uf), col]= df[col].median()
```

# Encoding Categorical Columns

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oheMnth= OneHotEncoder(sparse_output= False)
np_Mnth= oheMnth.fit_transform(df[["Month"]])
df_Mnth= pd.DataFrame(np_Mnth,columns= oheMnth.categories_ )
df_Mnth.drop('January', axis= 1, inplace=True)
```

```python
ohePay= OneHotEncoder(sparse_output= False)
np_Pay= ohePay.fit_transform(df[["Payment_Behaviour"]])
df_Pay= pd.DataFrame(np_Pay,columns= ohePay.categories_ )
df_Pay.drop('Low_spent_Small_value_payments', axis= 1, inplace=True)
```

# Splitting the Independent and Dependent Variables

```python
X= df.drop(['Month', 'Occupation', 'Payment_Behaviour', 'Credit_Score','index'], axis= 1)
y= df['Credit_Score']
```

```python
X= pd.concat([X, df_Mnth, df_Occu, df_Pay], axis=1)
```

```python
X
```

# Applying ANN Model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import Adam
```

```python
X.columns = X.columns.astype(str)
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
y = to_categorical(y, num_classes=3)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# Building the neural network model
model = Sequential()
model.add(Dense(125, input_dim=43, activation='relu', kernel_regularizer=l2(0.001)))
model.add(BatchNormalization())
model.add(Dropout(0.3))

model.add(Dense(100, activation='relu', kernel_regularizer=l2(0.001)))
model.add(BatchNormalization())
model.add(Dropout(0.3))

model.add(Dense(80, activation='relu', kernel_regularizer=l2(0.001)))
model.add(BatchNormalization())
model.add(Dropout(0.3))

model.add(Dense(50, activation='relu', kernel_regularizer=l2(0.001)))
model.add(BatchNormalization())
model.add(Dense(3, activation='softmax'))
```

```python
optimizer = Adam(learning_rate=0.0001)
model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
```

```python
checkpoint = ModelCheckpoint('best_model.keras', monitor='val_accuracy', save_best_only=True, verbose=1)
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
history = model.fit(X_train, y_train, epochs=300, batch_size=2000, verbose=1, validation_split=0.2,
                    callbacks=[checkpoint, early_stopping])
```

```python
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'][:25], label='train loss')
plt.plot(history.history['val_loss'][:25], label='val loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
model.load_weights('best_model.keras')
```

```python
y_pred_train = model.predict(X_train)
y_pred_train = np.argmax(y_pred_train, axis=1)
y_train = np.argmax(y_train, axis=1)

y_pred_test = model.predict(X_test)
y_pred_test = np.argmax(y_pred_test, axis=1)
y_test = np.argmax(y_test, axis=1)

# Accuracy
train_accuracy = accuracy_score(y_train, y_pred_train)
test_accuracy = accuracy_score(y_test, y_pred_test)

print(f"Train Accuracy: {train_accuracy}")
print(f"Test Accuracy: {test_accuracy}")
```
